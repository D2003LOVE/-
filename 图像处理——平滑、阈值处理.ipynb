### 图像平滑处理
图像平滑处理是为了减少噪声或细节，使图像变得更加柔和。平滑方法通常基于滤波器，例如高斯滤波和中值滤波。

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 读取图像
image = cv2.imread('data/00001.tiff',cv2.IMREAD_GRAYSCALE)

def show_image(title,img):
    plt.imshow(img,cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
show_image('s',image)
```

### 高斯滤波
高斯滤波是一种基于正态分布的滤波器，通过加权平均邻域像素的值来实现平滑.

在这里，cv2.GaussianBlur函数使用了一个5*5的卷积核和标准差为1的高斯函数来进行平滑，卷积大小越大，平滑效果越明显。

```python
# 高斯滤波，ksize=(5,5)为核大小，sigmax为标准差
gaussian_blur = cv2.GaussianBlur(image,(5,5),sigmaX=1)
# 显示高斯滤波效果
show_image("Gaussian Blur",gaussian_blur)
```

### 中值滤波
中值滤波是基于排序的滤波方法，通常用于去除椒盐噪声(salt and pepper noise).中值滤波的特定的是，它取邻域像素值的中位数作为新的像素值，能够更好地保留图像边缘。

```python
### 中值滤波，ksize=5为卷积核大小
median_blur = cv2.medianBlur(image,5)
# 显示中值滤波效果
show_image("Median Blur",median_blur)
```

### 图像阈值处理
图像阈值处理是将灰度图像转换为二值图像的过程。常见的阈值处理方法有固定阈值和自适应阈值。

#### 固定阈值处理

```python
# 固定阈值处理，阈值为127，最大值为255
_,binary_thresh = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
# 显示固定阈值处理效果
show_image('Binary Threshold',binary_thresh)
```

#### 自适应阈值处理
自适应阈值处理根据邻域像素的均值或高斯加权均值来计算阈值，适合处理光照不均匀的图像

```python
# 自适应阈值处理，采用均值阈值方法，区块大小为11，偏移量为2
adaptive_thresh_mean = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_MEAN_C,
                                            cv2.THRESH_BINARY,11,2)
# 显示自适应均值阈值效果
show_image('Adaptive Threshold Mean',adaptive_thresh_mean)
# 自适应阈值处理，采用高斯加权均值方法，区块大小为11，偏移量为2
adaptive_thresh_gaussian = cv2.adaptiveThreshold(image,255,
                                                  cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                                 cv2.THRESH_BINARY,11,2)
show_image('Adaptive Threshold Gaussian',adaptive_thresh_gaussian)
```
